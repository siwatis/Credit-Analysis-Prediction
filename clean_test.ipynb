```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/siwatis/Credit-Analysis-Prediction/main/loan_approval_dataset.csv')
```

```python
# Columns name: replace widespaces
df.columns = [col.replace(' ','') for col in df.columns]

# Drop loan_id column
df.set_index('loan_id', inplace=True)

# Set data type to each variables and clean space for categorical variable
target_col = ['loan_status']
categorical_col = ['no_of_dependents','education','self_employed','loan_term']
measure_col = ['income_annum','loan_amount','cibil_score','residential_assets_value','commercial_assets_value','luxury_assets_value','bank_asset_value']
for col in categorical_col:
    df[col] = df[col].astype(str).str.strip()
    df[col] = df[col].astype('category')
for col in measure_col:
    df[col] = df[col].astype(float)
```

```python
dummy_list = []
for col in df[categorical_col].columns:
    cat_name = col
    dummies = pd.get_dummies(df[col]).map(lambda x: 1 if x == True else 0)
    dummies.columns = cat_name + '_' + np.array(dummies.columns)
    dummy_list.append(dummies)
dummies = pd.concat(dummy_list, axis=1)
dummies.columns = [name.lower().replace(' ','_') for name in dummies.columns]
```

```python
pd.concat([df[measure_col],dummies], axis=1).corr()
```
